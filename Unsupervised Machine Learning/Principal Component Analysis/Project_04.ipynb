```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt 
from matplotlib import style
style.use('dark_background')
import pandas as pd
```

```python
X, y = make_blobs(n_samples = 2000, n_features = 5)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_rescaled = scaler.fit_transform(X)
```

# Selecting the Variance Ratio 

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.99)
reduced = pca.fit_transform(data_rescaled)
```

```python
reduced.shape
```

```python
pca_02 = PCA(n_components = 4)
X_pca = pca_02.fit_transform(data_rescaled)
```

```python
X_pca
```

```python
print(np.sum(pca_02.explained_variance_ratio_))
```
